a, Look for the missing values in all the columns and either impute them (replace with mean,
median, or mode) or drop them. Justify your action for this task.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("train.csv")
```

```python
# Drop 'New_Price' due to excessive missing values
df.drop('New_Price', axis=1, inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
print("Remaining missing values:\n", df.isnull().sum())
```

b, Remove the units from some of the attributes and only keep the numerical values (for
example remove kmpl from “Mileage”, CC from “Engine”, bhp from “Power”, and lakh from
“New_price”).

```python
df['Mileage'] = df['Mileage'].str.replace(r'\s+[a-zA-Z/]+', '', regex=True)
df['Engine'] = df['Engine'].str.replace(r'\s+CC', '', regex=True)
df['Power'] = df['Power'].str.replace(r'\s+bhp', '', regex=True)
```

```python
df.head()
```

```python
df['Mileage'] = pd.to_numeric(df['Mileage'], errors='coerce')
df['Engine'] = pd.to_numeric(df['Engine'], errors='coerce')
df['Power'] = pd.to_numeric(df['Power'], errors='coerce')

df['Mileage'].fillna(df['Mileage'].median(),inplace=True)
df['Engine'].fillna(df['Engine'].median(),inplace=True)
df['Power'].fillna(df['Power'].median(),inplace=True)
```

```python
print(df.isnull().sum())
```

c,  Change the categorical variables (“Fuel_Type” and “Transmission”) into numerical one hot
encoded value.

```python
fuel_mapping = {'Diesel': 0, 'Petrol': 1, 'Electric': 2}
df['Fuel_Type'] = df['Fuel_Type'].replace(fuel_mapping)

transmission_mapping = {'Manual': 0, 'Automatic': 1}
df['Transmission'] = df['Transmission'].replace(transmission_mapping)
```

```python
df.head()
```

d, Create one more feature and add this column to the dataset (you can use mutate function in
R for this). For example, you can calculate the current age of the car by subtracting “Year” value
from the current year.

```python
from datetime import datetime
current_year = datetime.now().year

df['Car_Age'] = current_year - df['Year']
```

```python
print(df[['Year', 'Car_Age']].head())
```

e, Perform select, filter, rename, mutate, arrange and summarize with group by operations (or
their equivalent operations in python) on this dataset.

```python
# select
required_columns = ['Name', 'Mileage', 'Engine', 'Seats']
df_selected = df[required_columns]
df_selected.head()
```

```python
#filter
filtered_df = df[(df['Mileage'].astype(float) > 20) & (df['Engine'].astype(float) < 1500)]
filtered_df.head()
```

```python
#rename
renamed_df = df.rename(columns={'Power_per_Liter': 'Power_per_Fuel'})
renamed_df.head()
```

```python
#arrange
sorted_df = df.sort_values(by=['Car_Age', 'Mileage'], ascending=[True, False])
sorted_df.head()
```

```python
#summarize
grouped_df = df.groupby('Fuel_Type').agg({'Mileage': 'mean'}).sort_values(by='Mileage', ascending=False)
grouped_df.head()
```

```python
summary_csv_file = 'train.csv'
grouped_df.to_csv(summary_csv_file, index=False)
```
