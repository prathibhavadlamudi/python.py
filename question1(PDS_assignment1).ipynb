The **first stage** of the basic workflow is data collection, input, or
creation.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Create dataset
data = {
    "Height": [65.8, 71.5, 69.4, 68.2, 67.8, 68.7, 69.8, 70.1, 67.9, 66.8],
    "Weight": [112, 136, 153, 142, 144, 123, 141, 136, 112, 120],
    "Age": [30, 19, 45, 22, 29, 50, 51, 23, 17, 39],
    "Grip_strength": [30, 31, 29, 28, 24, 26, 22, 20, 19, 31],
    "Frailty": ["N", "N", "N", "Y", "Y", "N", "Y", "Y", "N", "N"]
}

df = pd.DataFrame(data)
```

The **second stage** involves processing or cleaning of the data
produced in the first stage.


---



```python
plt.figure(figsize=(8, 5))
plt.hist(df["Grip_strength"], bins=5, color="skyblue", edgecolor="black")
plt.xlabel("Grip Strength (kg)")
plt.ylabel("Frequency")
plt.title("Histogram of Grip Strength Distribution")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df["Age"], y=df["Grip_strength"], hue=df["Frailty"], palette={"N": "blue", "Y": "red"})
plt.xlabel("Age (Years)")
plt.ylabel("Grip Strength (kg)")
plt.title("Scatter Plot: Grip Strength vs Age")
plt.legend(title="Frailty")
plt.show()
```

```python
plt.figure(figsize=(7, 5))
sns.violinplot(x="Frailty", y="Grip_strength", data=df, palette={"N": "blue", "Y": "red"})
plt.xlabel("Frailty")
plt.ylabel("Grip Strength (kg)")
plt.title("Violin Plot: Grip Strength by Frailty")
plt.show()
```

The **third stage** is data analysis.


```python
# Save all figures
plt.figure()
df.groupby("Frailty")["Grip_strength"].mean().plot(kind="bar", color=["blue", "red"], edgecolor="black")
plt.xlabel("Frailty")
plt.ylabel("Average Grip Strength (kg)")
plt.title("Bar Chart: Average Grip Strength by Frailty")
plt.xticks(rotation=0)
plt.savefig("results/figures/average_grip_strength.png")
plt.close()
```
