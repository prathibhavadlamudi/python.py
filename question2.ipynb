```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('StudentsPerformance.csv')
data.head()  # Display the first few rows
```

1. Scatter Plot: Math vs Writing Scores
python
Copy
Edit


```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='math score', y='writing score', data=data, color='purple')
plt.title('Math Scores vs Writing Scores')
plt.xlabel('Math Score')
plt.ylabel('Writing Score')
plt.show()
```

2. Count Plot: Test Preparation Course Enrollment

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='test preparation course', data=data, palette='pastel')
plt.title('Test Preparation Course Enrollment')
plt.xlabel('Test Preparation Course')
plt.ylabel('Count')
plt.show()
```

3. Bar Plot: Average Scores by Gender



```python
plt.figure(figsize=(8, 5))
data.groupby('gender')[['math score', 'reading score', 'writing score']].mean().plot(kind='bar', figsize=(8, 5), colormap='coolwarm')
plt.title('Average Scores by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Score')
plt.show()
```

4. Pie Chart: Distribution of Lunch Type

```python
plt.figure(figsize=(8, 5))
data['lunch'].value_counts().plot.pie(autopct='%1.1f%%', colors=['lightblue', 'lightcoral'])
plt.title('Distribution of Lunch Type')
plt.ylabel('')
plt.show()
```

 5. KDE Plot: Distribution of Reading Scores by Gender

```python
plt.figure(figsize=(8, 5))
sns.kdeplot(data[data['gender'] == 'male']['reading score'], label='Male', shade=True, color='blue')
sns.kdeplot(data[data['gender'] == 'female']['reading score'], label='Female', shade=True, color='red')
plt.title('Distribution of Reading Scores by Gender')
plt.xlabel('Reading Score')
plt.ylabel('Density')
plt.legend()
plt.show()
```
